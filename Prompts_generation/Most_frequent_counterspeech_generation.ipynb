```python
import pandas as pd
import numpy as np
import math
```

```python
df= pd.read_csv("./../counterspeech classification/prompts_generation_data.csv")
```

```python
df.dropna(inplace=True)
```

```python
df["label"].value_counts()
```

```python
# change label here to get the most frequent beginnings for that cs type
data=df[df["label"]=="humor"]

data.reset_index(inplace=True,drop=True)
map_positive_tone={}
for cs in data["processed_counterspeech"]:
    cs_start=""
    if(len(cs.split())<4):
        continue
    for i in range(4):
        cs_start+= cs.split()[i]+" "                        # first four words
    if cs_start in map_positive_tone.keys():
        map_positive_tone[cs_start]+=1
    else:
        map_positive_tone[cs_start]=1
sorted_map_positive_tone = sorted(map_positive_tone.items(), key=lambda x:x[1], reverse=True)
sorted_map_positive_tone
```
